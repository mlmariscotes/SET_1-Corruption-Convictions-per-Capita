```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('/content/drive/MyDrive/jpy/Set 1_Postgress #1.csv')
```

```python
df.head(10)
```

# 1. Create a scatter plot of state income vs. corruption convictions per capita.

```python
# scatter plot of total state income vs. corruption convictions per capita.
plt.figure(figsize=(10,6))
plt.scatter(df['total_income_state'], df['conviction_per_capita'], marker='x', color='red')
plt.title('Total State Income vs. Conviction per Capita')
plt.xlabel('State Income')
plt.ylabel('Convictions per Capita')
plt.grid(True)
plt.show()
```

# 2. Test whether state income and corruption convictions per capita are correlated. Use the df.corr() method for this task.

```python
# state income and corruption convictions per capita are correlated
correlation = df[['total_income_state', 'conviction_per_capita']].corr()

correlation
```

### Analysis
The correlation coefficient between total_income_state and conviction_per_capita is approximately **-0.140**. This value indicates a weak negative correlation between the two variables. That is, as the total income in a state increases, the number of corruption convictions per capita tends to slightly decrease, but the relationship is not strong.

# 3. Create a histogram of the income distribution for all states.

```python
# histogram of the total income distribution for all states
plt.figure(figsize=(10,6))
plt.hist(df['total_income_state'], bins=20, color='green', edgecolor='black')
plt.title('Distribution of Total Income across All States')
plt.xlabel('Total Income')
plt.ylabel('Number of States')
plt.grid(True)
plt.show()
```

### Analysis
Here is the histogram showing the distribution of total income across all states.

The x-axis represents the total income while the y-axis represents the number of states that fall into each income bracket.

From the histogram, we can observe that the distribution is somewhat skewed to the right, with a larger number of states having lower total incomes and a smaller number of states having higher total incomes.
